```python
import pandas as pd
from pathlib import Path

# Usamos la ruta ya verificada (proyecto / data / raw / final)
ruta_final = proyecto / 'data' / 'raw' / 'final'
archivo = 'resenas_google_maps.csv'          # <-- nombre corregido

df = pd.read_csv(ruta_final / archivo)

print("✅ Dataset cargado")
print("Dimensiones:", df.shape)
print("Columnas:", df.columns.tolist())
df.head(3)
```

```python
# Columnas que queremos conservar (ajustá los nombres exactos si hace falta)
columnas_deseadas = [
    'review_text',
    'review_rating',
    'category',
    'reviews_link',
    'name',
    'review_datetime_utc',
    'review_questions_Travel group'
]

# Revisamos cuáles existen realmente en el DataFrame
columnas_presentes = [col for col in columnas_deseadas if col in df.columns]
columnas_faltantes = [col for col in columnas_deseadas if col not in df.columns]

if columnas_faltantes:
    print("⚠️ Columnas no encontradas en el dataset y que no se conservarán:")
    for col in columnas_faltantes:
        print(f"  - {col}")

# Filtramos el DataFrame para que solo tenga las columnas presentes
df = df[columnas_presentes].copy()

print("\n✅ Columnas conservadas:")
print(df.columns.tolist())
print(f"Dimensiones actuales: {df.shape}")
df.head(3)
```

```python
# Duplicados considerando TODAS las columnas
duplicados_completos = df.duplicated()
total_duplicados = duplicados_completos.sum()

print(f"Total de filas duplicadas exactas (todas las columnas): {total_duplicados}")
if total_duplicados > 0:
    print("\nPrimeros 5 duplicados:")
    display(df[duplicados_completos].head(5))
else:
    print("No se encontraron filas completamente duplicadas.")
```

```python
# Detectar filas donde TODAS las columnas son NaN o vacías
filas_totalmente_vacias = df.isnull().all(axis=1) | (df.astype(str).apply(lambda x: x.str.strip() == '', axis=1).all(axis=1))

total_filas_vacias = filas_totalmente_vacias.sum()

print(f"Total de filas completamente vacías: {total_filas_vacias}")

if total_filas_vacias > 0:
    print("\nÍndices de las filas vacías:")
    print(df[filas_totalmente_vacias].index.tolist())

    # Eliminar filas completamente vacías
    df = df[~filas_totalmente_vacias].copy()

    print(f"\n✅ Filas eliminadas. Nuevo tamaño del dataset: {df.shape}")
else:
    print("No se encontraron filas completamente vacías.")
```

```python
# Diccionario con el mapeo de nombres
mapeo_nombres = {
    'review_text': 'reseña',
    'review_rating': 'calificación',
    'category': 'tipo_lugar',
    'reviews_link': 'fuente',
    'name': 'nombre',
    'review_datetime_utc': 'fecha',
    'review_questions_Travel group': 'grupo_de_viaje'
}

# Renombramos solo las columnas que existen en el DataFrame
df.rename(columns={k: v for k, v in mapeo_nombres.items() if k in df.columns}, inplace=True)

print("✅ Columnas renombradas:")
print(df.columns.tolist())
df.head(3)
```

```python
# Usamos .map en lugar de .applymap (pandas >= 2.1.0)
df = df.map(lambda x: x.lower() if isinstance(x, str) else x)

print("✅ Todas las celdas de texto han sido convertidas a minúsculas.")
df.head(3)
```

```python
# Dividimos por el espacio y nos quedamos con la primera parte (la fecha)
df['fecha'] = df['fecha'].str.split(' ').str[0]

print("✅ Fechas limpiadas. Ejemplos:")
df[['fecha']].head(10)
```

```python
# Condición: reseña es nula, vacía o contiene solo espacios
condicion_vacia = df['reseña'].isna() | (df['reseña'].str.strip() == '')

total_vacias = condicion_vacia.sum()
print(f"Total de reseñas vacías: {total_vacias}")

if total_vacias > 0:
    # Mostrar algunos ejemplos de las filas que se eliminarán (si hay)
    print("\nPrimeras 5 filas con reseña vacía:")
    display(df[condicion_vacia].head(5))

    # Eliminar esas filas
    df = df[~condicion_vacia].copy()

    print(f"\n✅ Filas eliminadas. Nuevo tamaño del dataset: {df.shape}")
else:
    print("No se encontraron reseñas vacías.")
```

```python
from langdetect import detect, LangDetectException
import numpy as np

def detectar_espanol(texto):
    try:
        return detect(texto) == 'es'
    except LangDetectException:
        # Si falla (por texto muy corto, solo números, etc.), lo consideramos no español
        return False

# Aplicamos la detección a la columna 'reseña'
print("Detectando idiomas... (puede tardar unos segundos)")
es_espanol = df['reseña'].apply(detectar_espanol)

# Contar resultados
total_no_espanol = (~es_espanol).sum()
print(f"Reseñas que NO están en español: {total_no_espanol}")

if total_no_espanol > 0:
    print("\nEjemplos de reseñas eliminadas (primeras 5):")
    display(df[~es_espanol][['reseña']].head(5))

# Filtrar solo las que son español
df = df[es_espanol].copy()

print(f"\n✅ Reseñas en español conservadas. Nuevo tamaño: {df.shape}")
```

```python
import re
def eliminar_emojis(texto):
    if not isinstance(texto, str):
        return ""
    # Elimina emojis y símbolos no latinos, conservando letras acentuadas, puntuación básica, espacios
    texto_limpio = re.sub(r'[^\w\s.,!?;:áéíóúüñÁÉÍÓÚÜÑ]', '', texto, flags=re.UNICODE)
    # Normalizar espacios múltiples
    texto_limpio = re.sub(r'\s+', ' ', texto_limpio).strip()
    return texto_limpio
```

```python
import re

def limpiar_caracteres_especiales(texto):
    """
    Conserva solo:
      - Letras (incluyendo acentos, diéresis y ñ/Ñ)
      - Números (0-9)
      - Espacios
    Elimina todo lo demás: puntuación, símbolos, etc.
    """
    if not isinstance(texto, str):
        return ""

    # Mantiene solo letras (incluye tildes y ñ), números y espacios
    texto = re.sub(r'[^a-záéíóúüñA-ZÁÉÍÓÚÜÑ0-9\s]', '', texto, flags=re.UNICODE)

    # Normaliza espacios múltiples y recorta extremos
    texto = re.sub(r'\s+', ' ', texto).strip()

    return texto

# Aplicar la limpieza
df['reseña'] = df['reseña'].apply(limpiar_caracteres_especiales)

# Verificar si alguna reseña quedó vacía después de la limpieza
vacias_post_limpieza = (df['reseña'] == '') | df['reseña'].isna()
total_vacias_post = vacias_post_limpieza.sum()
print(f"Reseñas vacías después de la limpieza: {total_vacias_post}")

if total_vacias_post > 0:
    df = df[~vacias_post_limpieza].copy()
    print(f"✅ Filas eliminadas. Nuevo tamaño: {df.shape}")

# Mostrar ejemplos
print("\nEjemplos de reseñas completamente limpias:")
df[['reseña']].head(5)
```

```python
from spellchecker import SpellChecker
import pandas as pd

# Crear el corrector en español
spell = SpellChecker(language='es')

def contar_errores(texto):
    """Cuenta las palabras que no están en el diccionario."""
    if not isinstance(texto, str) or texto.strip() == '':
        return 0
    palabras = texto.split()
    # Palabras que el corrector considera erróneas
    desconocidas = spell.unknown(palabras)
    return len(desconocidas)

def palabras_erroneas(texto):
    """Devuelve una lista de las palabras probablemente mal escritas."""
    if not isinstance(texto, str) or texto.strip() == '':
        return []
    palabras = texto.split()
    desconocidas = spell.unknown(palabras)
    return list(desconocidas)

# Aplicar a cada reseña
print("Analizando errores (puede tardar unos segundos)...")
df['cantidad_errores'] = df['reseña'].apply(contar_errores)
df['palabras_erroneas'] = df['reseña'].apply(palabras_erroneas)

# Resumen estadístico
total_resenas = len(df)
resenas_con_errores = (df['cantidad_errores'] > 0).sum()
print(f"\nTotal de reseñas: {total_resenas}")
print(f"Reseñas con al menos un posible error: {resenas_con_errores} ({100*resenas_con_errores/total_resenas:.1f}%)")

# Mostrar algunas reseñas con más errores
top_erroneas = df.nlargest(10, 'cantidad_errores')[['reseña', 'cantidad_errores', 'palabras_erroneas']]
print("\n🔟 Las 10 reseñas con más posibles errores ortográficos:")
for i, row in top_erroneas.iterrows():
    print(f"\nReseña (errores: {row['cantidad_errores']}):")
    print(row['reseña'][:200] + ('...' if len(row['reseña']) > 200 else ''))
    print(f"Palabras dudosas: {row['palabras_erroneas'][:15]}...")
```

Los errores más comunes que cometen las personas a la hora de dejar su reseña en el sitio de google maps son:
1. Escribir las palabras pegadas haciendo que no tenga sentido el texto sin la debida separación
2. Tener bastantes errores ortográficos (confundiendo u omitiendo letras)
3. Repitiendo las mismas palabras una al lado de la otra
